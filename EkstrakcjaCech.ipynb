# Dyskretyzacja zmiennej ciągłej

```python
df = pd.DataFrame(data={'height': [175., 178.5, 185., 191., 184.5, 183., 168.]})
df
```

```python
# zmienną ciągłą chcę przedstawić jako zmienną kategoryczną

df['height_cut'] = pd.cut(x=df.height, bins=3)

# przekazuje kolumnę na której chcę zrobić podział (height) i bins=3 to na ile podziałów chcę podzielić moją ciongłą
```

```python
df
```

```python
# gdy chcę mieć konkretne przydziały, w bins podaje konkretne wartości jako tuple []

df['height_cut'] = pd.cut(x=df.height, bins=[160,175,180, 195])
df
```

```python
# mogę też wyświetlić kolejną etykietę

df['height_cut'] = pd.cut(x=df.height, bins=[160,175,180,195], labels=['small', 'medium', 'high'])
df

# pozwala nam wyświetlić konkretną wartość tekstową
```

```python
# jak mam to co wyżej to mogę użyć get_dummies

pd.get_dummies(df, drop_first=True, prefix='height').astype('int')
```

# Ekstrakcja cech

```python
df = pd.DataFrame(data={'lang': [['PL', 'ENG'], ['GER', 'ENG', 'PL', 'FRA'], ['RUS']]})
df
```

```python
# chcę mieć kolumnę z liczbą języków

df['lang_number'] = df['lang'].apply(len)
df
```

```python
# buduje flagę - czy aplikacja jest napisana także w języku polskim

# funkca lambda zwróci 1 jeżeli 'PL' jest w x czyli elemencie który przekazujemy

df['PL_flag'] = df['lang'].apply(lambda x: 1 if 'PL' in x else 0)
df
```

```python
df = pd.DataFrame(data={'website': ['wp.pl', 'onet.pl', 'google.com']})
df
```

```python
# rozdzielenie stringów w różna kolumny

# bibliotek pandas pozdala nam rozdzielać wyrazy po użyciu metod tekstowych str.split
# dzielę tekst względem kropki a dzieki expand = True mogę zbudować dwie nowe kolumny

df.website.str.split('.', expand=True)
```

```python
# jak to przypisać do nowego DataFrama?

new = df.website.str.split('.', expand=True)
df['portal'] = new[0]
df['extension'] = new[1]
df
```
