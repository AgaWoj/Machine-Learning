# Dyskretyzacja zmiennej cigej

```python
df = pd.DataFrame(data={'height': [175., 178.5, 185., 191., 184.5, 183., 168.]})
df
```

```python
# zmienn cig chc przedstawi jako zmienn kategoryczn

df['height_cut'] = pd.cut(x=df.height, bins=3)

# przekazuje kolumn na ktrej chc zrobi podzia (height) i bins=3 to na ile podziaw chc podzieli moj ciong
```

```python
df
```

```python
# gdy chc mie konkretne przydziay, w bins podaje konkretne wartoci jako tuple []

df['height_cut'] = pd.cut(x=df.height, bins=[160,175,180, 195])
df
```

```python
# mog te wywietli kolejn etykiet

df['height_cut'] = pd.cut(x=df.height, bins=[160,175,180,195], labels=['small', 'medium', 'high'])
df

# pozwala nam wywietli konkretn warto tekstow
```

```python
# jak mam to co wyej to mog uy get_dummies

pd.get_dummies(df, drop_first=True, prefix='height').astype('int')
```

# Ekstrakcja cech

```python
df = pd.DataFrame(data={'lang': [['PL', 'ENG'], ['GER', 'ENG', 'PL', 'FRA'], ['RUS']]})
df
```

```python
# chc mie kolumn z liczb jzykw

df['lang_number'] = df['lang'].apply(len)
df
```

```python
# buduje flag - czy aplikacja jest napisana take w jzyku polskim

# funkca lambda zwrci 1 jeeli 'PL' jest w x czyli elemencie ktry przekazujemy

df['PL_flag'] = df['lang'].apply(lambda x: 1 if 'PL' in x else 0)
df
```

```python
df = pd.DataFrame(data={'website': ['wp.pl', 'onet.pl', 'google.com']})
df
```

```python
# rozdzielenie stringw w rna kolumny

# bibliotek pandas pozdala nam rozdziela wyrazy po uyciu metod tekstowych str.split
# dziel tekst wzgldem kropki a dzieki expand = True mog zbudowa dwie nowe kolumny

df.website.str.split('.', expand=True)
```

```python
# jak to przypisa do nowego DataFrama?

new = df.website.str.split('.', expand=True)
df['portal'] = new[0]
df['extension'] = new[1]
df
```
